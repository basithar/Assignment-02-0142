```python

#Q1
import cv2 as cv
import numpy as np

# Read image in grayscale
img = cv.imread(r"C:\Users\User\Pictures\Screenshots\a1.png", cv.IMREAD_GRAYSCALE)

# Apply Canny Edge Detection
edges = cv.Canny(img, 550, 690)

# Resize images (width, height)
img_resized = cv.resize(img, (500, 400))
edges_resized = cv.resize(edges, (500, 400))

cv.imshow('Original Image (Resized)', img_resized)
cv.imshow('Canny Edge Image (Resized)', edges_resized)

cv.waitKey(0)
cv.destroyAllWindows()

```

```python

#Q2
from matplotlib import pyplot as plt
# Extract coordinates of edge points
indices = np.where(edges != 0)
x = indices[1]
y = indices[0]

# Scatter plot
plt.figure(figsize=(6, 6))
plt.scatter(x, y, s=5, c='black')
plt.title('Scatter Plot of Edge Points')
plt.xlabel('x')
plt.ylabel('y')
plt.gca().invert_yaxis()   # Makes image-like coordinate view
plt.grid(True)
plt.show()

```

```python

#Q3
# Least Squares Fit using polyfit
m_lsf, c_lsf = np.polyfit(x, y, 1)

# Generate fitted line
x_line = np.linspace(np.min(x), np.max(x), 500)
y_line_lsf = m_lsf * x_line + c_lsf

# Plot
plt.figure(figsize=(6, 6))
plt.scatter(x, y, s=5, c='black', label='Edge Points')
plt.plot(x_line, y_line_lsf, 'r', linewidth=2, label='Least Squares Fit')
plt.title('Least Squares Fit')
plt.xlabel('x')
plt.ylabel('y')
plt.gca().invert_yaxis()
plt.legend()
plt.grid(True)
plt.show()

print("Least Squares slope (m) =", m_lsf)
print("Least Squares intercept (c) =", c_lsf)

```

```python
#Q6
import numpy as np
import matplotlib.pyplot as plt

# 1. Prepare the data
# Assuming 'x' and 'y' are already defined from your edge detection 
# If not, ensure 'indices = np.where(edges != 0)' is run first
points = np.column_stack((x, y)) 
mean = np.mean(points, axis=0) 
centered = points - mean 

# 2. Covariance matrix & Eigen decomposition
cov = np.cov(centered.T) 
eigenvalues, eigenvectors = np.linalg.eig(cov) 

# 3. Find the direction of the largest eigenvalue
# This represents the principal component (the line of best fit)
direction = eigenvectors[:, np.argmax(eigenvalues)] 
dx, dy = direction[0], direction[1] 

# 4. Create the TLS line coordinates
# t defines the length of the line in both directions from the mean
t = np.linspace(-300, 300, 500) 
x_tls = mean[0] + t * dx 
y_tls = mean[1] + t * dy 

# 5. Plotting
plt.figure(figsize=(6, 6)) 

# Scatter plot of the original edge points
plt.scatter(x, y, s=5, c='black', label='Edge Points') 

# Total Least Squares Fit line
plt.plot(x_tls, y_tls, 'b', linewidth=2, label='Total Least Squares Fit') 

# Formatting the plot
plt.title('Total Least Squares Fit') 
plt.xlabel('x') 
plt.ylabel('y') 

# Invert Y-axis because image coordinates (0,0) start at the top-left
plt.gca().invert_yaxis() 

plt.legend() 
plt.grid(True) 
plt.show()
```
